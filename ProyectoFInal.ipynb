```python
import pandas as pd
#Base de datos

url='https://raw.githubusercontent.com/Daviddsal/ProyectoRealidadVirtual/refs/heads/main/Proyecto_limpio_1.3.csv'
df=pd.read_csv(url)
df
```

# **Gráfcas básicas de cada columna**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Crear el histograma y obtener los datos de frecuencias
plt.figure(figsize=(8, 6))  # Define el tamaño del gráfico
n, bins, patches = plt.hist(df['Edad'], bins=30, color='lightgreen', edgecolor='black')  # Generar histograma

# Identificar el índice del bin con frecuencia máxima
max_freq_index = np.argmax(n)  # Índice del bin con la frecuencia más alta

# Identificar el índice del bin con la frecuencia más baja mayor que 0
min_freq_index = np.where(n > 0)[0][np.argmin(n[n > 0])]  # Índice del bin con frecuencia mínima mayor que 0

# Resaltar los bins correspondientes
for i, patch in enumerate(patches):
    if i == max_freq_index:  # Frecuencia máxima
        patch.set_facecolor('green')  # Resaltar en rojo
    elif i == min_freq_index:  # Frecuencia mínima mayor a 0
        patch.set_facecolor('green')  # Resaltar en azul

# Personalizar el gráfico
plt.title('Histograma de las Edades', fontsize=16)  # Título del gráfico
plt.xlabel('Edades', fontsize=10)  # Etiqueta del eje x
plt.ylabel('', fontsize=10)  # Etiqueta del eje y

plt.show()
```

```python
# Crear el boxplot
plt.figure(figsize=(8, 3))
plt.boxplot(df['Edad'], patch_artist=True, notch=True, boxprops=dict(facecolor='lightgreen', color='black'), vert=False)

# Personalizar el gráfico
plt.title('Boxplot de edad', fontsize=16)
plt.ylabel('', fontsize=12)

# Mostrar el gráfico
plt.show()
```

# **Gráficos de correlación**

```python
import pandas as pd
#Base de datos completamente numérica

url='https://raw.githubusercontent.com/Daviddsal/ProyectoRealidadVirtual/refs/heads/main/Proyecto_limpio_num%C3%A9rico.csv'
df2=pd.read_csv(url)
df2
```

```python
df3 = df2.drop(columns=['Student_ID'])

# Calcular la matriz de correlación
correlation_matrix = df3.corr()
correlation_matrix
```

```python
plt.figure(figsize=(20,20))
# Gráfico de calor
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')

# Mostrar gráfico
plt.show()
```

```python
# Reducimos nuestros rango
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', vmin=-0.5, vmax=0.5)
```

```python
# Pairplot para visualizar la relación entre las variables numéricas
sns.pairplot(df3)
# Mostrar gráfico
plt.title('Pairplot de Variables Numéricas')
plt.show()
```

```python
# Crear un boxplot para cada columna numérica
num_cols = df2.select_dtypes(include=['number']).columns  # Filtrar columnas numéricas

plt.figure(figsize=(30, 100))  # Ajustar tamaño general del gráfico

for i, col in enumerate(num_cols, start=1):
    plt.subplot(len(num_cols), 1, i)  # Crear subgráficos para cada columna
    plt.boxplot(df2[col].dropna(), patch_artist=True, notch=True,
                boxprops=dict(facecolor='lightblue', color='black'), vert=False)

    # Personalizar cada gráfico
    plt.title(f'Boxplot de {col}', fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.tight_layout()  # Evitar superposición de gráficos

# Mostrar todos los gráficos
plt.show()
```

```python
# Crear el mapa de calor de valores faltantes
plt.figure(figsize=(12, 8))  # Ajustar tamaño del gráfico
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)

# Personalizar el gráfico
plt.title('Mapa de Calor de Valores Faltantes', fontsize=16)
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Filas', fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
